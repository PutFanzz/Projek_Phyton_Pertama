```
from IPython.display import HTML, display

# Kod animasi grafik menggunakan HTML5 + JavaScript untuk paparan lancar di Colab
game_canvas_html = """
<div id="game-container" style="text-align: center; font-family: 'Courier New', Courier, monospace; background-color: #1a1a1a; padding: 20px; border-radius: 10px; width: 520px; margin: auto;">
    <h2 style="color: #00e676; margin-bottom: 10px;">🚀 GALACTIC RANGER 🚀</h2>
    <div style="color: white; margin-bottom: 10px; font-size: 18px;">SCORE: <span id="score-val">0</span></div>
    <canvas id="spaceCanvas" width="500" height="300" style="border: 3px solid #00e676; background-color: #050505;"></canvas>
    <p style="color: #aaa; font-size: 13px; margin-top: 10px;">Klik kotak game & guna <b>Arrow Up / Down</b> untuk gerak.</p>
    <button id="restart-btn" style="padding: 8px 15px; background-color: #00e676; color: black; font-weight: bold; border: none; border-radius: 5px; cursor: pointer; display: none;" onclick="resetGame()">Main Semula</button>
</div>

<script>
    const canvas = document.getElementById("spaceCanvas");
    const ctx = canvas.getContext("2d");
    const scoreDisplay = document.getElementById("score-val");
    const restartBtn = document.getElementById("restart-btn");

    // Tetapan Kapal Angkasa
    let shipY = 135;
    let shipX = 30;
    let shipSize = 20;
    let shipSpeed = 8;

    // Tetapan Meteor
    let meteors = [];
    let maxMeteors = 7;
    let meteorSpeed = 4;

    let score = 0;
    let gameInterval;
    let gameActive = false;
    let keysPressed = {};

    function startGame() {
        if (!gameActive && restartBtn.style.display === "none") {
            gameInterval = setInterval(updateGame, 1000 / 60); // Jalankan pada 60 FPS (Animasi Lancar)
            gameActive = true;
        }
    }

    function updateGame() {
        // Logik Pergerakan Kapal (Berdasarkan key yang ditekan)
        if (keysPressed["ArrowUp"] && shipY > 10) {
            shipY -= shipSpeed;
        }
        if (keysPressed["ArrowDown"] && shipY < canvas.height - shipSize - 10) {
            shipY += shipSpeed;
        }

        // Penjanaan Meteor Secara Rawak
        if (meteors.length < maxMeteors && Math.random() < 0.05) {
            meteors.push({
                x: canvas.width,
                y: Math.random() * (canvas.height - 20) + 10,
                radius: Math.random() * 8 + 6
            });
        }

        // Pergerakan & Semakan Perlanggaran Meteor
        for (let i = meteors.length - 1; i >= 0; i--) {
            meteors[i].x -= meteorSpeed;

            // Semak jarak antara kapal dan meteor (Collision Detection)
            let distX = Math.abs(meteors[i].x - (shipX + shipSize/2));
            let distY = Math.abs(meteors[i].y - (shipY + shipSize/2));

            if (distX < (shipSize/2 + meteors[i].radius) && distY < (shipSize/2 + meteors[i].radius)) {
                return gameOver();
            }

            // Buang meteor yang terlepas dan tambah skor
            if (meteors[i].x + meteors[i].radius < 0) {
                meteors.splice(i, 1);
                score += 10;
                scoreDisplay.innerText = score;

                // Setiap 100 mata, lajukan sikit meteor
                if (score % 100 === 0) {
                    meteorSpeed += 0.5;
                }
            }
        }

        draw();
    }

    function draw() {
        // Bersihkan skrin lama
        ctx.fillStyle = "#050505";
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        // Lukis Bintang Latar Belakang (Hiasan Animasi)
        ctx.fillStyle = "#ffffff";
        for(let i=0; i<10; i++) {
            let starX = (Math.random() * canvas.width);
            let starY = (Math.random() * canvas.height);
            ctx.fillRect(starX, starY, 1, 1);
        }

        // Lukis Kapal Angkasa (Bentuk Segitiga Neon)
        ctx.fillStyle = "#00e676";
        ctx.beginPath();
        ctx.moveTo(shipX + shipSize, shipY + shipSize/2);
        ctx.lineTo(shipX, shipY);
        ctx.lineTo(shipX, shipY + shipSize);
        ctx.closePath();
        ctx.fill();

        // Lukis Meteor (Bulatan Merah/Oren)
        ctx.fillStyle = "#ff5252";
        for (let m of meteors) {
            ctx.beginPath();
            ctx.arc(m.x, m.y, m.radius, 0, Math.PI * 2);
            ctx.fill();
        }
    }

    function gameOver() {
        clearInterval(gameInterval);
        gameActive = false;

        ctx.fillStyle = "rgba(0, 0, 0, 0.8)";
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        ctx.fillStyle = "#ff5252";
        ctx.font = "bold 25px Courier New";
        ctx.textAlign = "center";
        ctx.fillText("MISI GAGAL / GAME OVER", canvas.width / 2, canvas.height / 2);

        restartBtn.style.display = "inline-block";
    }

    function resetGame() {
        shipY = 135;
        meteors = [];
        score = 0;
        meteorSpeed = 4;
        scoreDisplay.innerText = score;
        restartBtn.style.display = "none";
        startGame();
    }

    // Pendengar Event Keyboard
    window.addEventListener("keydown", function(e) {
        if (["ArrowUp", "ArrowDown"].indexOf(e.key) > -1) {
            e.preventDefault(); // Elak skrin Google Colab daripada skrol naik turun
        }
        keysPressed[e.key] = true;

        if (!gameActive && restartBtn.style.display === "none") {
            startGame();
        }
    });

    window.addEventListener("keyup", function(e) {
        keysPressed[e.key] = false;
    });

    // Render paparan awal
    draw();
</script>
"""

# Paparkan kotak game terus dalam output cell Colab
display(HTML(game_canvas_html))
```
